```python
import pandas as pd

# Load the dataset
file_path = "../data/Online_Retail.xlsx"  # Update with your file path
df = pd.read_excel(file_path)

# Check the dataset structure
print(f"Initial dataset size: {df.shape}")
print(df['Country'].value_counts())  # View row distribution by country

# Sampling configuration
total_rows_to_keep = 7000 # Target size after deletion
countries = df['Country'].unique()
rows_per_country = total_rows_to_keep // len(countries)  # Approx rows to keep per country

# Stratified sampling
sampled_df = df.groupby('Country').apply(lambda x: x.sample(n=min(len(x), rows_per_country), random_state=42))
sampled_df = sampled_df.reset_index(drop=True)  # Reset index after grouping

# Check the reduced dataset
print(f"Reduced dataset size: {sampled_df.shape}")
print(sampled_df['Country'].value_counts())  # Verify diversity

# Save the reduced dataset
output_path = "../data/final_dataset.xlsx"  # Update with your desired output path
sampled_df.to_excel(output_path, index=False)
print(f"Reduced dataset saved to: {output_path}")
```
